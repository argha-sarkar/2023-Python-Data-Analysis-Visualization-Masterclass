## Part 2

* The `mount_everest_deaths` dataset has its own index column provided in the dataset.  When importing it, use the existing index column.
* Which columns have zero null values?
* Which column has the most null values?


```python
#The mount_everest_deaths dataset has its own index column provided in the dataset. When importing it, use the existing index column
mou = pd.read_csv('data\\mount_everest_deaths.csv', index_col=0)
mou.head()
```

```python
mou.isnull().sum()
```

```python
#Which columns have zero null values?
for i in mou.columns:
    if mou[i].isnull().sum() == 0:
        print(i)
```

```python
#Which column has the most null values?
max_null_count = 0
max_null_column = ''

for i in mou.columns:
    null_count = mou[i].isnull().sum()
    if null_count > max_null_count:
        max_null_count = null_count
        max_null_column = i
        
print(f"{max_null_column} {max_null_count}")
```
